Ques1   Create a numpy array x containing data representing time(in hrs) spend on studing : [1,2,3,4,5]
                Create a  numpy array y containing corresponding data representing study score(out of 100) : [50,60,70,80,90]
                Plot a line graph with blue solid line
                Overlay (on top of it) a scatter plot with red markers

```python
#pip install matplotlib

import matplotlib.pyplot as plt
import numpy as np
# x is student studying hours
x=np.array([1,2,3,4,5])
# y is out of 100 marks
y=np.array([50,60,70,80,90])
plt.plot(x,y)
plt.show()
```

```python
#part2
import matplotlib.pyplot as plt
import numpy as np
# x is student studying hours
x=np.array([1,2,3,4,5])
# y is out of 100 marks
y=np.array([50,60,70,80,90])
plt.plot(x,y)
plt.scatter(x,y,color='r')
plt.show()
```

# 3.Plot a bar graph represeting subject and its score
# 4.Color each bar differently
# 5.Add label,title,and modify the grid lines to display (--)style

```python
#part1
import matplotlib.pyplot as plt
import numpy as np
sub=np.array(['Maths','Science', 'History','English','Arts'])
score=np.array([85,70,75,90,80])
plt.bar(sub,score)
plt.show()
```

```python
#part2
import matplotlib.pyplot as plt
import numpy as np
sub=np.array(['Maths','Science', 'History','English','Arts'])
score=np.array([85,70,75,90,80])
mycolors=np.array(["red","green","blue","yellow","pink"])
plt.bar(sub,score, color=mycolors)
plt.show()
```

```python
#part3
import matplotlib.pyplot as plt
import numpy as np
sub=np.array(['Maths','Science', 'History','English','Arts'])
score=np.array([85,70,75,90,80])
plt.title(" Score of the students subjectwise of class CBP")
plt.xlabel("Subjects of the students")
plt.ylabel("Score of the Students")
plt.grid(color = 'black',linestyle = '--')
mycolors=np.array(["red","green","blue","yellow","pink"])
plt.bar(sub,score, color=mycolors)
plt.show()
```
